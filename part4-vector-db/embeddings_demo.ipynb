```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler delivered a fast yorker",
    "The cricket team won the tournament",
    "The bowler took three wickets in one over",

    # Cooking
    "I added spices while cooking the curry",
    "Baking a cake requires precise measurements",
    "The chef prepared a delicious pasta dish",

    # Cybersecurity
    "Hackers attempted to breach the system",
    "Strong passwords improve cybersecurity",
    "Data encryption protects sensitive information"
]
```

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(sentences)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

top_indices = similarities.argsort()[-2:][::-1]

for idx in top_indices:
    print("Sentence:", sentences[idx])
    print("Similarity Score:", similarities[idx])
    print()
```
